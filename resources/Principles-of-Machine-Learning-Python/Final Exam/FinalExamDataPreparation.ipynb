### Further prepare data

* Convert birthday to age
* Drop irrelevant, non-categorizable cells such as name or title

```python
def prepare_data(data_frame):
    prepared_data = data_frame
    prepared_data['BirthdayDate'] =  pd.to_datetime(prepared_data['BirthDate'], infer_datetime_format=True)
    prepared_data['Age'] = np.floor((pd.Timestamp(day=31, month=12, year=1998)-prepared_data['BirthdayDate'])/np.timedelta64(1, 'Y'))    
    prepared_data = prepared_data.drop(columns=['Title', 'CustomerID', 'FirstName', 'MiddleName', 'LastName', 'Suffix', 'AddressLine1', 'AddressLine2', 'City', 'BirthDate', 'StateProvinceName', 'PostalCode', 'PhoneNumber', 'BirthdayDate'])
    return prepared_data

prep_train = prepare_data(training_data)
prep_test = prepare_data(aw_test)

prep_train.head(5)
```

### One hot encode categorigal data - normalize real data

```python
data_grid = prep_train
test_grid = prep_test

# --- One hot encode all categorical features ---
def encode_string(cat_features, test_features):
    ## First encode the strings to numeric categories
    enc = preprocessing.LabelEncoder()
    enc.fit(cat_features)
    enc_cat_features = enc.transform(cat_features)
    enc_tr_features = enc.transform(test_features)
    ## Now, apply one hot encoding
    ohe = preprocessing.OneHotEncoder(categories='auto')
    encoded = ohe.fit(enc_cat_features.reshape(-1,1))
    return encoded.transform(enc_cat_features.reshape(-1,1)).toarray(), encoded.transform(enc_tr_features.reshape(-1,1)).toarray()

first_column = 'CountryRegionName'
categorical_columns = ['Education', 'Occupation', 'Gender', 'MaritalStatus']
Features, Test_Features = encode_string(data_grid[first_column], test_grid[first_column])
for col in categorical_columns:
    temp, temp2 = encode_string(data_grid[col], test_grid[col])
    Features = np.concatenate([Features, temp], axis = 1)
    Test_Features = np.concatenate([Test_Features, temp2], axis = 1)
    
categorical_feature_count = Features.shape[1]
print("{} categorical features".format(categorical_feature_count))
    
Features = np.concatenate([Features, np.array(data_grid[['HomeOwnerFlag', 'NumberCarsOwned', 
                            'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome', 'Age']])], axis = 1)    
Test_Features = np.concatenate([Test_Features, np.array(test_grid[['HomeOwnerFlag', 'NumberCarsOwned', 
                            'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome', 'Age']])], axis = 1)

# --- Normalize all continuous features ---
scaler = preprocessing.StandardScaler().fit(Features[:,categorical_feature_count:])
Features[:,categorical_feature_count:] = scaler.transform(Features[:,categorical_feature_count:])
Test_Features[:,categorical_feature_count:] = scaler.transform(Test_Features[:,categorical_feature_count:])

# -- Normalize spent amount
spend_np_array_org = spend_np_array = np.log(np.array(spend[:]).reshape(-1,1))
spend_scaler = preprocessing.StandardScaler().fit(spend_np_array)
spend_np_array = spend_scaler.transform(spend_np_array)
print(spend_scaler.scale_)
print(spend_scaler.mean_)
print(spend_scaler.var_)

restored_scaler = preprocessing.StandardScaler()
restored_scaler.mean_ = spend_scaler.mean_
restored_scaler.scale_ = spend_scaler.scale_
restored_scaler.var_ = spend_scaler.var_
spend_np_array_verify = spend_scaler.transform(spend_np_array_org)

spend_scaler_config = pd.DataFrame(columns=['mean','scale','variance'])
spend_scaler_config.loc[0] = [spend_scaler.mean_[0], spend_scaler.scale_[0], spend_scaler.var_[0]]
spend_scaler_config.to_csv("ExportedScalerConfig.csv", index=False)

print(np.equal(spend_np_array_verify, spend_np_array).sum())
```
